### Project Name: Loan Approval Classification

Dataset: https://www.kaggle.com/datasets/anishdevedward/loan-approval-dataset?resource=download

The dataset is in csv format and includes information about people who applied for loans. The dataset includes a person's name, city of residence, credit score, annual income, loan amount, number of years employed, calculated risk factor, and whether or not they were approved for a loan.

We will try to predict whether or not given values, a person would be approved for a loan or not

### Potential Implementation Issues

One feature we may need to drop from the table is a person's name as this really has no impact on a person's ability to be approved for a loan. We also may need to go through the dataset to see if a city of residence has an impact on a persons ability to be approved for a loan or not. If there's no correlation, we may be able to drop that attribute as well. 

### Stakeholders/Results

The results of this classification are useful because it may become a tool for people who are interested in applying for loans to see if they have a good chance of being approved. While our model may not be fully accurate, it should give people a good idea about whether or not they will be approved. People who could find these result set beneficial are people who are thinking about applying for loans, and people who give out loans. This model may be useful for money lenders because they can use it as an initial weed out tool to move people along in the system before manually going in and checking whether or not to approve someone.
