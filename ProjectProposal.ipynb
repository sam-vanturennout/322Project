### Project Name

Loan Approval Classification

### Dataset Description:

We will use the Loan Approval Dataset from Kaggle (CSV).
Source: https://www.kaggle.com/datasets/anishdevedward/loan-approval-dataset

The dataset contains loan applications with attributes such as applicant name, city, credit score, annual income, loan amount, years employed, calculated risk factor, and whether the loan was approved.

* Class attribute: Loan_Approved (approved vs. not approved).
* Features we will use: City, Credit Score, Annual Income, Loan Amount, Years Employed, Risk Factor.
* Features to drop: Name (identifier, no predictive value).

### Implementation / Technical Plan

We will perform:

* EDA: Summaries, distributions, class balance.
* Preprocessing: Drop non-informative attributes, encode City, check for missing values, scale numeric features if needed.
* Models: Baseline (majority class) + Decision Tree, k-NN, and/or Logistic Regression.
* Evaluation: Train/test split or cross-validation; accuracy and precision/recall.

### Anticipated Challenges

* Class imbalance: May require stratified splits and comparison with majority baseline.
* Uninformative attributes: City may add noise; we will test performance with and without it.
* Missing or inconsistent values: Handle by imputation or dropping affected rows.
* Correlated features: Risk Factor may duplicate information; we will check correlations.

**Feature Selection:**
Use correlations, mutual information, and tree-based feature importance to confirm which attributes matter. Remove unhelpful city categories if needed. 

### Potential Impact and Stakeholders

A loan-approval predictor can help applicants estimate approval likelihood and help lenders quickly screen applications before manual review. The results may also highlight which factors influence approval decisions most.

Stakeholders: loan applicants, financial institutions, and (indirectly) regulators concerned with fairness.
